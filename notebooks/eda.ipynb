# Time for PANDAS
- "Pandas" has a reference to both "Panel Data", and "Python Data Analysis"

```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/Titanic-Dataset.csv")
df.head(500)
```

```python
# df.dropna(inplace=True) 
# Dropping reduces the number of rows drastically so cannot drop all null values
print(df.info())
```

```python
new_df = df.fillna(120)

new_df.info()
```

```python
x = df["Age"].median()
cab= df["Cabin"].mode()[0]
emb = df["Embarked"].mode()[0]
df.fillna({"Age": x}, inplace=True)
df.fillna({"Cabin": cab}, inplace=True)
df.fillna({"Embarked": emb}, inplace=True)
df.info()
df.head()
```

**- loc is label-based indexing in pandas**

```python
# print(df.duplicated().to_string())
df.drop_duplicates(inplace= True)
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
corr_matrix = df.corr(numeric_only=True)

print(corr_matrix)
```

```python
df.plot(kind='scatter', x = "Age", y = "Survived")
plt.show()
```
