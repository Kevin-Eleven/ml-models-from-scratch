```python
import numpy as np

# 1. GENERATE DATA
# We will use smaller numbers or normalize them to prevent overflow
x_raw = np.array([[i] for i in range(0, 500, 2)])
y_raw = np.array([[i * i] for i in range(0, 500, 2)])

# 2. DATA NORMALIZATION (Crucial step!)
# Divide by max value to keep inputs between 0 and 1
x = x_raw / np.max(x_raw)
y = y_raw / np.max(y_raw)

# Hyperparameters
lr = 0.1  # Can use larger LR because data is normalized
epochs = 1000

# Parameters
w = 0.0
b = 0.0

for i in range(epochs):
    # Forward Pass
    y_pr = w * x + b

    # Calculate Difference (NOT Squared Error) for gradient
    diff = y - y_pr

    # Calculate Gradients
    # Derivative of MSE wrt w is: -2 * average( error * x )
    dw = -2 * np.average(diff * x)
    db = -2 * np.average(diff)

    # Update Weights
    w = w - (lr * dw)
    b = b - (lr * db)

    # Optional: Print progress
    if i % 100 == 0:
        # Calculate actual MSE for display
        mse = np.mean((y - y_pr) ** 2)
        print(f"Epoch {i}: Loss {mse:.5f}")

print(f"Final w: {w}, Final b: {b}")
```

```python
import matplotlib.pyplot as plt

# 1. Calculate predictions using your trained weights
# (Make sure 'w' and 'b' are the final values from your training loop)
y_pred = w * x + b

# 2. Create the plot
plt.figure(figsize=(8, 5))  # Optional: Make the figure larger

# Plot the ACTUAL data as blue dots
# 'label' helps create the legend later
plt.scatter(x, y, color="blue", label="Actual Data (Curve)")

# Plot the PREDICTED model as a red line
plt.plot(x, y_pred, color="red", linewidth=3, label="Predicted Line")

# 3. Add labels for clarity (Best Practice)
plt.title("Linear Regression on Quadratic Data")
# plt.xlabel("Input (x)")
plt.ylabel("Output (y)")
plt.legend()  # Shows the labels defined above

# 4. Show the plot
plt.show()
```
